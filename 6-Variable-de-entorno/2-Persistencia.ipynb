# Persistencia de las variables de entorno
Como vimos anteriormente, las variables de entorno locales y globales están "activas" dentro de una sesión del intérprete, donde fueron definidas.

Para asegurar la persistencia de las variables, es decir, que estén disponibles para usarlas en cualquier usuario y cualquier sesión del intérprete, podemos hacer lo siguiente.

## 1. Definir las variables en el archivo ```.bashrc```
El archivo ```.bashrc``` se encuentra en nuestra carpeta raíz de cada usuario, por lo que la variable estará disponible para el usuario que la creó.

````bash
ls -al
````

Dentro del archivo se definen procesos y servicios de arranque:

````bash
pwd
````

````bash
cd
````

````bash
pwd
````

````bash
cat .bashrc
````

Utilizar la terminal para modificar este archivo y agregar lo siguiente:
```bash
export HOLAVARBASHRC = "Hola desde la variable global en el archivo .bashrc"
```

Accedemos a la variable:

````bash
pwd
````

Volvemos a cargar el archivo que acabamos de modificar:

````bash
source .bashrc
````

````bash
echo $HOLAVARBASHRC
````

## 2. Modificar el archivo ```environment``` (root)
si queremos ejecutar la variable ```HOLAVARBASHRC``` desde el usuario root no será accesible:



Para poder definir la variable dentro del aechivo ```environment``` debemos cambiar al usuario root:
```bash
sudo su -
```
Dentro del usuario root vamos a:
```
cd /etc#
```
Luego:
```bash
nano environment
```

Con ```nano environment``` abrimos el editor y agregamos lo siguiente:
```bash
HOLAVARENVIRONMENT="Hola desde el archivo enviroment con el usuario root"
```

Salimos del usuario root con ```exit``` y volvemos a entrar para comprobar el archivo **env** que ahora contiene la variable creada:

````bash
cd ../../..
````

````bash
pwd
````

````bash
source /etc/environment
````

````bash
echo $HOLAVARENVIRONMENT
````

La variable es accesible desde nuestro usuario normal aunque fue creada desde root.

## Conclusiones
Si creamos la variable en el archivo **.bashrc** con nuestro usuario normal, la variable será accesible desde este mismo usuario y desde cualquier directorio. Si creamos la variable en el archivo **environment** con nuestro usuario root, la variables estará disponible desde cualquier lugar.
